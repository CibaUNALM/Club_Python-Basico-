**El orden lexicogr√°fico**: es el orden en el que se colocan las palabras en un diccionario. En Python, el orden lexicogr√°fico se aplica a los objetos de tipo string, que se comparan car√°cter por car√°cter. Si un car√°cter en una cadena es menor que otro en otra cadena, la primera cadena es considerada menor que la segunda.

```python
print("apple" < "banana")
print("cat" < "dog") 
print("egg" < "eggs")

print("Zebra" < "apple") 
print("apple" < "Apple")
```

El orden lexicogr√°fico en Python es sensible a may√∫sculas y min√∫sculas, por lo que las cadenas que contienen letras may√∫sculas se colocan antes que las que contienen solo letras min√∫sculas. Es importante tener en cuenta que en Python, el orden lexicogr√°fico se aplica a todos los tipos de secuencias, incluyendo listas, tuplas y conjuntos, y no solo a las cadenas.

# 7. FUNCIONES

##  7.1. BREAK (ruptura)

```python
# Salga del bucle cuando sea "banana":x
fruits = ["apple", "banana", "cherry", "pear"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```python
# Salir del bucle cuando es "pl√°tano", Pero esta vez el descanso llega antes de la impresi√≥n:x
fruits = ["apple", "banana", "cherry", "pear"]
for x in fruits:
  if x == "banana":
    break
  print(x)
```

##  7.2. CONTINUE

```python
# Con la declaraci√≥n continue podemos detener el iteraci√≥n actual del bucle y contin√∫e con la siguiente
fruits = ["apple", "banana", "cherry", "pear"]
for x in fruits:
  if x == "banana": # No imprime pl√°tano
    continue
  print(x)
```

La funci√≥n *__ str __()*

```python
class persona:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def __str__(self):
    return f"{self.name}({self.age})"

p2 = persona("John", 36)
print(p2)
p2
```

## 8.2.1. M√âTODOS DE OBJETOS

```python
class persona:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def myfunc(self):
    print("Hello my name is " + self.name)

p1 = persona("John", 36)
p1.myfunc()
```

El par√°metro *self*

Es una referencia a la instancia actual de la clase, y se utiliza para acceder a las variables que pertenecen a la clase.self

```python
class persona:
  def __init__(mysillyobject, name, age): # mysillyobject en lugar de self
    mysillyobject.name = name
    mysillyobject.age = age
    
  def myfunc(abc):                        # abc en lugar de self
    print("Hello my name is " + abc.name)

p1 = persona("John", 36)
p1.myfunc()
```
