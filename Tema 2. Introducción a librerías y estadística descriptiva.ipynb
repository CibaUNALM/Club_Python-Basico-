### np.nan

```python
# Crea una matriz con algunos valores faltantes (np.nan)
matriz = np.array([[1, 2, np.nan], [4, np.nan, 6], [np.nan, 8, 9]])
print("Matriz con valores faltantes: \n", matriz)
```

```python
# Sustituye los valores faltantes (np.nan) por un valor específico (0)
matriz_sustituida = np.where(np.isnan(matriz), 0, matriz)
print("Matriz con valores faltantes sustituídos: \n", matriz_sustituida)
```

```python
# Calcula la media de cada columna de la matriz (ignora los valores faltantes)
media_columna = np.nanmean(matriz, axis=0)
print("Media de cada columna: \n", media_columna)
```

```python
# Calcula la suma de cada fila de la matriz (ignora los valores faltantes)
suma_fila = np.nansum(matriz, axis=1)
print("Suma de cada fila: \n", suma_fila)
```

### Creación de un Dataframe

```python
# Crea un DataFrame a partir de un diccionario
data = {
    'Producto': ['Leche', 'Pan', 'Jugo', 'Agua'],
    'Precio': [3.0, 4.5, 2.0, 1.5]
}

df = pd.DataFrame(data)
print("DataFrame: \n", df)
```

### Filtros

```python
data1 = {
    'Producto': ['Leche', 'Pan', 'Jugo', 'Agua', 'Manzanas', 'Peras'],
    'Precio': [3.0, 4.5, 2.0, 1.5, 2.0, 2.5],
    'Cantidad Vendida': [20, 15, 10, 25, 30, 20]
}
df1 = pd.DataFrame(data1)
print("DataFrame original: \n", df1)
```

```python
# Utiliza una condición para filtrar los datos
df_filtered = df1[df1['Precio'] > 2.5]
print("DataFrame filtrado: \n", df_filtered)
```

```python
# Utiliza varias condiciones para filtrar los datos (usando & y |)
df_filtered = df1[(df1['Precio'] > 2.5) & (df1['Cantidad Vendida'] >= 20)]
print("DataFrame filtrado con varias condiciones: \n", df_filtered)
```

```python
# Utiliza isin para filtrar los datos
df_filtered = df1[df1['Producto'].isin(['Leche', 'Manzanas'])]
print("DataFrame filtrado con isin: \n", df_filtered)
```

### Agrupación

```python
# Agrupa los datos en el DataFrame
grouped = df1.groupby('Producto').mean()
print("DataFrame agrupado: \n", grouped)
```

### Unión de Dataframes

```python
# Crea el primer DataFrame
data1 = {'Producto': ['Agua', 'Jugo', 'Leche'],
         'Precio': [2.5, 3.0, 4.0]}
df1 = pd.DataFrame(data1)
print("DataFrame 1: \n", df1)
```

```python
# Crea el segundo DataFrame
data2 = {'Producto': ['Pan', 'Galletas', 'Leche'],
         'Precio': [5.5, 2.0, 4.0]}
df2 = pd.DataFrame(data2)
print("DataFrame 2: \n", df2)
```

```python
# Une los dos DataFrames con pd.concat()
concatenado = pd.concat([df1, df2], ignore_index=True)
print("DataFrames concatenados con pd.concat(): \n", concatenado)
```

```python
# Une los dos DataFrames con pd.merge()
mergeado = pd.merge(df1, df2, on='Producto', how='outer')
print("DataFrames mergeados con pd.merge(): \n", mergeado)
```

La función **pd.concat()** une los DataFrames en un solo DataFrame, mientras que **pd.merge()** une los DataFrames basándose en una columna específica (en este caso, la columna 'Producto').

# Medidas de Tendencia Central

### Media arimetica

La media aritmética es la suma de todos los valores de un conjunto de datos dividida por su número de elementos. Se puede calcular en Python utilizando la función mean de Numpy.

```python
import numpy as np

datos = [1, 2, 3, 4, 5]
media = np.mean(datos)
print("Media aritmética:", media)
```

### Mediana

La mediana es el valor central de un conjunto de datos ordenado. Se puede calcular en Python utilizando la función median de Numpy.

```python
import numpy as np

datos = [1, 2, 3, 4, 5]
mediana = np.median(datos)
print("Mediana:", mediana)
```

### Moda

La moda es el valor que ocurre con mayor frecuencia en un conjunto de datos. Se puede calcular en Python utilizando la función mode de la librería scipy.

```python
from scipy import stats

datos = [1, 2, 3, 3, 4, 5]
moda = stats.mode(datos)
print("Moda:", moda.mode[0])
```

# Medidas de Dispersión.


Las medidas de dispersión se utilizan para describir la variabilidad de un conjunto de datos. 

### Desviación estándar

La desviación estándar es una medida de la variabilidad de un conjunto de datos con respecto a su media. Se puede calcular en Python utilizando la función std de Numpy.

```python
import numpy as np

datos = [1, 2, 3, 4, 5]
desviacion = np.std(datos)
print("Desviación estándar:", desviacion)
```

### Rango

El rango es la diferencia entre el valor máximo y el valor mínimo de un conjunto de datos. Se puede calcular en Python utilizando las funciones max y min de Numpy.

```python
import numpy as np

datos = [1, 2, 3, 4, 5]
rango = np.max(datos) - np.min(datos)
print("Rango:", rango)
```

### Varianza

La varianza es una medida de la variabilidad de un conjunto de datos con respecto a su media. Se puede calcular en Python utilizando la función var de Numpy.

```python
import numpy as np

datos = [1, 2, 3, 4, 5]
varianza = np.var(datos)
print("Varianza:", varianza)
```

# Percentiles, Deciles y Cuartiles

Los percentiles son una medida que divide un conjunto de datos ordenados en 100 partes iguales e indica el porcentaje de datos que se encuentra debajo de su valor.

```python
import numpy as np

datos = [1,2,3,4,5,6,7,8,9]
p10 = np.percentile(datos,10)
p50 = np.percentile(datos,50)
p90 = np.percentile(datos,90)

#Otra forma 
P10 = np.quantile(datos, 0.1)
P50 = np.quantile(datos, 0.5)
P90 = np.quantile(datos, 0.9)

print("Percentil 10:", p10, " ",P10)
print("Percentil 50:", p50, " ",P50)
print("Percentil 90:", p90, " ",P90)
```

```python
from statistics import quantiles
datos = [1,2,3,4,5,6,7,8,9]
#Percentiles
p = quantiles(datos, n=100)

#Cuartiles
q = quantiles(datos, n=4)

#Deciles
d = quantiles(datos, n=10)

print("Percentil 10:", p[9])
print("Percentil 50:", p[49])
print("Percentil 90:", p[89])
print('\n')
print("Cuartil 1:", q[0])
print("Cuartil 2:", q[1])
print("Cuartil 3:", q[2])
print('\n')
print("Decil 1:", d[0])
print("Decil 5:", d[4])
print("Decil 9:", d[8])
```

# Correlación

La correlación es una medida que establece la relación lineal entre dos variables numéricas. Siempre toma un valor entre -1 y 1 donde:

*  -1 indica una correlación lineal perfectamente negativa entre dos variables
*  0 indica que no hay correlación lineal entre dos variables
*  1 indica una correlación lineal perfectamente positiva entre dos variables

```python
import pandas as pd
datos = {
    'A': [8,4,9,1,8],
    'B': [0,1,6,8,0],
    'C': [9,7,8,1,8]
}

datos = pd.DataFrame(datos)
```

```python
datos.corr()
```

```python
datos['A'].corr(datos['B'])
```

```python
datos['A'].corr(datos['B'], method='spearman')
```

```python
import numpy as np
np.corrcoef(datos['A'], datos['B'])
```

```python
import numpy as np
np.corrcoef(datos['A'], datos['B'])[0,1]
```
